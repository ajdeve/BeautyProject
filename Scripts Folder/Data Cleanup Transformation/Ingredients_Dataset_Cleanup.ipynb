```python
import pandas as pd 
```

```python
#This is to clean up product data and ingredient data.
#We want to make two tables.
#one table contains the product id and the ingredient id it contains.
#The second table is the ingredient id and ingredients
```

```python
#read csv
sephora = pd.read_csv('sephora_website_dataset-unique.csv')
ingredients = pd. read_csv("ingredients.csv")
```

```python
#Separate each ingredient: the ingredients of the product are put together and connected with "-"
sephora["ingredients"] = sephora["ingredients"].str.split('-')
```

```python
#Rearrange the product table so that each row has only one ingredient
split_s = sephora.explode('ingredients')
split_s
```

```python
#Match the hazardous characteristics of each ingredient
leftjoin=pd.merge(split_s,ingredients,left_on='ingredients',right_on='chemical_name',how='left')
leftjoin.head()
```

```python
#Only keep the ingredients and its hazad traits.
harmful_ingredients = leftjoin[["ingredients", "hazard_traits"]]
```

```python
#There are some same ingredients in different products, so delete duplicate ingredients
harmful_ingredients = harmful_ingredients.drop_duplicates('ingredients', keep='first')
```

```python
#Change the name of the columns
harmful_ingredients = harmful_ingredients.rename(columns={ "ingredients": 'chemical_name'})
harmful_ingredients = harmful_ingredients.rename(columns={ "hazard_traits": 'hazard_trait'})

#Add id to each ingredient
harmful_ingredients.insert(0, 'ingredient_id', harmful_ingredients.index + 1)
harmful_ingredients
```

```python
# Cleanup ingredient name
harmful_ingredients['chemical_name'] = harmful_ingredients['chemical_name'].str.replace('*', '', regex = False)
```

```python
#Change the hazard trait of non-hazardous ingredients to "NULL"
harmful_ingredients=harmful_ingredients.fillna("NULL")
harmful_ingredients
```

```python
# download csv
harmful_ingredients.to_csv('harmful_ingredients.csv', sep=',',index=False)
```

```python
#Product id and their ingredients
product_ingredients = leftjoin[["id", "ingredients"]]
product_ingredients
```

```python
#Match ingredients and their id to connect product id and ingredient id
product_ingredients=pd.merge(product_ingredients,harmful_ingredients,left_on='ingredients',right_on='chemical_name',how='left')
product_ingredients
```

```python
#Only keep product id and ingredient id
product_ingredients = product_ingredients[["id", "ingredient_id"]]
product_ingredients = product_ingredients.rename(columns={ "id": 'product_id'})
product_ingredients
```

```python
product_ingredients.to_csv('product_ingredients.csv', sep=',',index=False)
```
